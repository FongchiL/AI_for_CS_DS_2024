```python
import pandas as pd
import requests

# Set up API endpoint and API key
endpoint = 'https://leetcode.com/api/problems/all/'
headers = {'Authorization': 'Bearer <your LeetCode API key here>'}

# Send API request and get JSON response
response = requests.get(endpoint, headers=headers)
data = response.json()

# Convert JSON response to pandas DataFrame
df = pd.json_normalize(data['stat_status_pairs'])

# Drop unnecessary columns
#df = df.drop(columns=['status', 'paid_only', 'difficulty.level', 'frequency'])

# Rename columns
df = df.rename(columns={
    'stat.frontend_question_id': 'id',
    'stat.question__title': 'title',
    'stat.question__title_slug': 'slug',
    'stat.question__hide': 'is_hidden',
    'stat.total_acs': 'total_accepted',
    'stat.total_submitted': 'total_submissions',
    'difficulty.rating': 'difficulty',
})

# Save DataFrame to CSV file
df.to_csv('leetcode_questions_all.csv', index=False)
```
