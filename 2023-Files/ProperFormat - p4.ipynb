```python

import pandas as pd
import openai
import numpy as np
df = pd.read_csv("Function_Declarations.csv")
df
```

```python
# gets all of the empty cells
arr = np.where(pd.isnull(df["Answer_Java"]))
empty_list = list(arr)
```

```python
def contains_comments(code):
    for i in range(len(code)-1):
        if code[i] == '/' and code[i + 1] == '/':
            return True
    return False
```

```python
import openai
def helper(cell, i, template):
    repeat = 0
    openai.api_key = ""
    
    while True: 
        if(repeat == 30):
            print("skipping question")
#             skipped.append(i)
            return
        print("making api call")
        completion = openai.ChatCompletion.create(
            #model="gpt-4",
            model="gpt-3.5-turbo",
            messages=[
                {"role": "user", "content": cell}
            ]
        )
            
        if template == completion.choices[0].message.content:
            print("gave template again.")
            repeat+=1
            continue
        if completion.choices[0].message.content[:5] in ["class", "SELEC", "UPDAT", "impor"] or completion.choices[0].message.content[:3] in ["cat"] or completion.choices[0].message.content[:4] in ["grep"] or completion.choices[0].message.content[:3] in ["/**"]:
            if completion.choices[0].message.content[:3] in ["/**"]:
                break
            elif contains_comments(completion.choices[0].message.content) == False:
                break
            else:
                print(completion.choices[0].message.content)
                ex = completion.choices[0].message.content
                print("contains comments")
        else:
            print("incorrect answer format")
            repeat+=1
#             print(completion.choices[0].message.content)
            
    #adding to data frame
    df.loc[i, 'Correct_Format_Answer'] = completion.choices[0].message.content
    
    #adding to csv
    df.to_csv("Final-p4.csv", index=False)
```

```python
count = 1200
```

```python
for i, row in df.loc[count:].iterrows():
    if count == 1600:
        break
    count = i
    print(count)
    if count in empty_list[0]:
        continue
        
    while True:        
        try:        
            helper("Can you rewrite this java code: " + str(row["Answer_Java"]) + "\n" + "\n" + "\n" + "And rewrite it to this format:" + str(row["code_snippet"]) + 
                   ". Just provide the code, and no explanation along with no comments.", i, str(row["code_snippet"]))
            break
        except:
            print("Rate Limit Error")
print("finished")
```
