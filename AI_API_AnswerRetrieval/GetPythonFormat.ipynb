```python
import subprocess
import pandas as pd
import csv
```

```python
contents = []
```

```python
def get_python_file_path(id, slug):
    id = int(id)
    new_id = f"{id:04d}"
    return "python/" + new_id + "." + slug + "/solution.py"
```

```python
def add_declaration_csv(csv_path):
    with open(csv_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            id_value = row['id']
            slug = row['slug']
            if id_value == '':
                break
            print(id_value)
            python_file_path = get_python_file_path(id_value, slug)
            try:
                with open(python_file_path, 'r') as file:
                    content = file.read()
                contents.append(content)
            except FileNotFoundError:
                print('File was not found')
                contents.append('Python3 not supported')
            
```

```python
add_declaration_csv('answers_python2024.csv')
```

```python
df = pd.read_csv('answers_python2024.csv')
df = df.drop('Answer', axis=1)
df['Python_Declaration'] = contents
df
```

```python
df.to_csv('pythondeclarations.csv', index=False) 
```
