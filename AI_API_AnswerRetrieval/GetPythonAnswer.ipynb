```python
def getDeclaration(s):
    decl = ''
    inDec = False
    for i in range(len(s)):
        if inDec and s[i] == ':':
            inDec = False 
        if inDec:
            decl += s[i]
        if s[i] == 'd' and s[i+1] == 'e' and s[i+2] == 'f':
            decl += s[i]
            inDec = True
    return decl
```
