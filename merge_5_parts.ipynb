```python
import pandas as pd
```

```python
p1 = pd.read_csv("Final-p1.csv")
p1 = p1[0:400]
p1
```

```python
p2 = pd.read_csv("Final-p2.csv")
p2 = p2[400:800]
p2
```

```python
p3 = pd.read_csv("Final-p3.csv")
p3 = p3[800:1200]
p3
```

```python
p4 = pd.read_csv("Final-p4.csv")
p4 = p4[1200:1600]
p4
```

```python
p5 = pd.read_csv("Final-p5.csv")
p5 = p5[1600:]
p5
```

```python
import numpy as np
merged = pd.concat([p1, p2, p3, p4, p5], axis = 0)
merged["Correct_Format_Answer"].replace('', np.nan, inplace=True)
merged.dropna(subset=["Correct_Format_Answer"], inplace = True)
merged
```

```python
merged.to_csv("Merged.csv", index=False)
```
