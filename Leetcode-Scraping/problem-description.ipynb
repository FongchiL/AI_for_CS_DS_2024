```python
import pandas as pd
df = pd.read_csv("leetcode_questions_all.csv")
df
```

```python
import requests
from bs4 import BeautifulSoup
import time
session = requests.Session()
probDescList = []
solutionList = []
likesList = []
dislikesList = []
for index, row in df.iterrows():
    url = 'https://leetcode.com/problems/' + row['slug'] + '/'
    print()
    print("*******************"+ url +"**********************")
    print()
    while True:
        try:
            response = session.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')
            problem_desc = soup.find('meta', property="og:description").get("content")
            divs = soup.find_all("div", {"class": "text-xs"})
            break
        except AttributeError:
            print("HANGING")
            #time.sleep(10)
            continue
            
    solutions = divs[2].text
    try:
        likes = divs[5].text
    except IndexError:
        likes = "premium"
    try:
        dislikes = divs[6].text
    except IndexError:
        dislikes = "premium"
    
    
    probDescList.append(problem_desc)
    solutionList.append(solutions)
    likesList.append(likes)
    dislikesList.append(dislikes)
#     print(problem_desc)
#     print(solutions)
#     print(likes)
#     print(dislikes)
    
```

```python
# df["ProblemDescription"] = probDescList
df["num_solutions"] = solutionList
df["num_likes"] = likesList
df["num_dislikes"] = dislikesList
df
```

```python
df.to_csv('leetcode_questions_all_excluding_prem.csv', index=False)
```
